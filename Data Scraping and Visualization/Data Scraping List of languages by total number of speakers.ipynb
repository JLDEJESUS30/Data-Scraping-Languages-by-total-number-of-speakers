# List of languages by total number of speakers

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_languages_by_total_number_of_speakers"
website = requests.get(url)
soup = BeautifulSoup(website.text, 'html')    
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'static-row-numbers plainrowheaders vertical-align-top sticky-header-multi sort-under sortable wikitable jquery-tablesorter')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
language = table.find_all('th')
```

```python
language
```

```python
total_speakers = [title.text.strip() for title in language]

print(total_speakers)
```

```python
df = pd.DataFrame(columns = total_speakers)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df.index = range(1, len(df) + 1)
df
```
